```python
from pyspark.sql import functions as F

tables = [
    "silver_students",
    "silver_assessments",
    "silver_lms_events",
    "silver_vle_activity",
    "silver_instructors",
    "gold_course_difficulty",
    "gold_engagement",
    "gold_instructor_effectiveness",
    "gold_student_risk",
    "gold_student_risk_predictions",
    "gold_weekly_trends",
]

rows = []

for t in tables:
    full_name = f"dbo.{t}"
    print(f"\n{'='*90}")
    print(f"TABLE: {full_name}")
    print('='*90)

    try:
        df = spark.table(full_name)
        df.printSchema()

        cols = df.columns
        row_count = df.count()

        print(f"\nColumns ({len(cols)}):")
        for c in cols:
            print(f"- {c}")

        rows.append((t, len(cols), row_count, ", ".join(cols)))

    except Exception as e:
        print(f"ERROR reading {full_name}: {e}")
        rows.append((t, None, None, f"ERROR: {e}"))

summary_df = spark.createDataFrame(
    rows,
    ["table_name", "column_count", "row_count", "columns"]
)

display(summary_df)
```

```python
audit_rows = []

for t in tables:
    full_name = f"dbo.{t}"
    try:
        df = spark.table(full_name)
        for field in df.schema.fields:
            audit_rows.append((
                t,
                field.name,
                field.dataType.simpleString(),
                field.nullable
            ))
    except Exception as e:
        audit_rows.append((t, "__ERROR__", str(e), False))

audit_df = spark.createDataFrame(
    audit_rows,
    ["table_name", "column_name", "data_type", "nullable"]
)

display(audit_df)
```
